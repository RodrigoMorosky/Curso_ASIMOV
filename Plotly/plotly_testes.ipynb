```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def consulta_bc(codigo_bcb):
  url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json'.format(codigo_bcb)
  df = pd.read_json(url)
  df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')
  #df.set_index('data', inplace=True)
  return df
```

```python
ipca = consulta_bc(433)
```

```python
display(ipca)
```

```python
ipca_2 = ipca[ipca['data']>='2000-01-01']
```

```python
igpm = consulta_bc(189)
```

```python
igpm_2 = igpm[igpm['data']>='2000-01-01']
```

```python
selic_meta = consulta_bc(432)
```

```python
selic_meta_2 = selic_meta[selic_meta['data']>='2000-01-01']
```

```python
cdi = consulta_bc(12)
```

```python
cdi_2 = cdi[cdi['data']>='2000-01-01']
```

```python
ipca_2 = ipca['data']>'2000-01-01'
```

```python
fig = go.Figure()
fig.add_traces(go.Scatter(x=ipca['data'],y=ipca['valor'],mode='lines',name='IPCA'))
fig.add_traces(go.Scatter(x=igpm['data'],y=igpm['valor'],mode='lines',name='IGPM'))
fig.show()
```

```python
fig = go.Figure()
fig.add_traces(go.Scatter(x=ipca_2['data'],y=ipca_2['valor'],mode='lines',name='IPCA'))
fig.add_traces(go.Scatter(x=igpm_2['data'],y=igpm_2['valor'],mode='lines',name='IGPM'))
fig.show()
```

```python
fig = make_subplots(rows=2,cols=2)
fig.add_trace(go.Scatter(x=ipca_2['data'],y=ipca_2['valor'],mode='lines',name='IPCA'),row=1,col=1)
fig.add_trace(go.Scatter(x=igpm_2['data'],y=igpm_2['valor'],mode='lines',name='IGPM'),row=1,col=2)
fig.add_trace(go.Scatter(x=selic_meta_2['data'],y=selic_meta_2['valor'],mode='lines',name='Selic Meta'),row=2,col=1)
fig.add_trace(go.Scatter(x=cdi_2['data'],y=cdi_2['valor'],mode='lines',name='CDI'),row=2,col=2)
fig.update_layout(height=600)
fig.show()
```
